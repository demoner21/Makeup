# Introduction: Image Background Removal and Display

This Python code utilizes the **´rembg´** library and the Python Imaging Library (**´PIL´**) to perform background removal from an image and visually display the before-and-after results. The primary functionality is encapsulated within two functions: **remove_background** and **display_images_side_by_side**.

### 1. Background Removal Function: remove_background

- This function takes as input the path of an image file, the desired output path for the processed image, and optional parameters for specifying the width and height of the output.
- It opens the input image using PIL, resizes it if width and height parameters are provided, and employs the rembg library to remove the background.
- The resulting image is resized to match the specified dimensions and saved at the specified output path.
- Additionally, the function calls **display_images_side_by_side** to visually compare the original and processed images side by side.

### 2. Display Function: display_images_side_by_side

- This function creates a new image with double width, pastes the input and output images side by side, and displays the composite image.

### 3. Execution Section

- The code sets the input file name (file_name) to 'Poetry.jpg', specifies the output file path (output_path) as 'output2.png', and defines fixed width and height (fixed_width and fixed_height) for resizing.
- The full input path is constructed, and the **remove_background** function is called with these parameters.
- The result is the removal of the image background, resizing of the images, saving of the processed image, and a visual display of the original and processed images side by side.

```python
from rembg import remove
from PIL import Image
import matplotlib.pyplot as plt
```

- **´rembg´** library: 
This library is used for background removal in images. It provides a function called remove that takes an image with a background and returns the same image with the background removed.

- **´PIL´ (Pillow)** library:
This is a powerful Python Imaging Library that adds image processing capabilities to your Python interpreter. In this context, it's likely being used for opening, manipulating, and saving images.

- **matplotlib.pyplot**: 
This library is used for creating visualizations, and we import the pyplot module from it with the alias plt. It will be used to display images.

```python
def remove_background(input_path, output_path, width=None, height=None):
    try:
        input_image = Image.open(input_path)

        # Resize if width and height are provided
        if width and height:
            input_image = input_image.resize((width, height))

        output_image = remove(input_image)
        output_image = output_image.resize((width, height))  # Resize output image to match

        # Save the output image
        output_image.save(output_path)
        print(f"Background removed successfully. Result saved at {output_path}")

        # Display input and output images side by side
        display_images_side_by_side(input_image, output_image)

    except Exception as e:
        print(f"Error: {e}")
```

- The function takes the input image path, output image path, and optional width and height parameters.
- It opens the input image using PIL and resizes it if width and height are provided.
- The remove function from the rembg library is used to remove the background from the input image.
- The output image is resized to match the specified width and height.
- The result is saved to the output path, and a success message is printed.
- The display_images_side_by_side function is assumed to be defined elsewhere, used for displaying input and output images side by side.
- Any errors during the process are caught and printed.

```python
def display_images_side_by_side(image1, image2):
    # Create a new blank image with double width
    new_image = Image.new('RGB', (image1.width + image2.width, image1.height))

    # Paste the input and output images side by side
    new_image.paste(image1, (0, 0))
    new_image.paste(image2, (image1.width, 0))

    # Display the side-by-side images
    new_image.show()

# Set the input file name
file_name = 'Poetry.jpg'

# Specify the output file path
output_path = 'output2.png'

# Fixed width and height
fixed_width, fixed_height = 300, 300

# Construct the full input path
input_path = file_name

# Remove background, resize, and display side-by-side images
remove_background(input_path, output_path, fixed_width, fixed_height)
```

- The code sets up the necessary libraries (**´rembg´** and **´PIL´**) and the **display_images_side_by_side** function.

- ´file_name´ is set to 'Poetry.jpg', specifying the input image file name.

- **´output_path´** is set to 'output2.png', specifying the output file path for the image with the removed background.

- **fixed_width** and **fixed_height** are set to 300, determining the width and height for resizing the images.

- The full **input_path** is constructed using the **file_name**.

- Finally, the **remove_background** function is called with the input and output paths, as well as the fixed width and height. This function removes the background, resizes the images, saves the result, and displays the input and output images side by side.

```python
# Load an image using PIL
image_path = output_path  # Replace with the actual path to your image file
image = Image.open(image_path)  # Open the image file using PIL's Image class

# Display the image using matplotlib.pyplot
plt.imshow(image)  # Use the imshow function to display the image
plt.axis('off')  # Turn off axis labels
plt.show()  # Show the image
```

- **image_path**: Replace this variable with the actual path to your image file.
- **Image.open()**: Uses the open method from the Image class to open the image file.
- **plt.imshow()**: Displays the image using the imshow function from matplotlib.pyplot.
- **plt.axis('off')**: Turns off axis labels for a cleaner image display.
- **plt.show()**: Shows the image.
